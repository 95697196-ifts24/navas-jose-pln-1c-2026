## 1. El problema: para la computadora, el texto todavia es texto plano

En esta primera parte vamos a mirar que pasa cuando tratamos un texto solo como una secuencia de caracteres.


```python
texto = "La CEO de ACME, Lisa Chang, viajara a Mexico para supervisar los nuevos proyectos de energia renovable. Ella se mostro muy entusiasmada."
print(texto)
```

Antes de seguir, hace una prueba breve:
- calcula `len(texto)`;
- compara ese valor con la salida de `texto.split()`;
- preguntate que informacion linguistica falta.

Idea central: separar por espacios no equivale a comprender el lenguaje.


```python
palabras = texto.split()
print(palabras)
```

### Por que necesitamos algo mejor

`split()` resuelve una parte del problema, pero deja preguntas abiertas:
- `"ACME,"` y `"Chang,"` conservan la coma. No son palabras limpias.
- `"La"` y `"los"` son articulos, mientras que `"viajara"` es una forma conjugada del verbo `"viajar"`. Como llevas cada palabra a su forma base para analizar frecuencias o temas?
- Como identifica la maquina que `"ACME"` es una organizacion y que `"Lisa Chang"` es una persona?

**Pausa breve de observacion:** si un programa solo ve caracteres, todavia no puede distinguir personas, organizaciones, lugares ni relaciones gramaticales.


## 2. La solucion: spaCy como interprete linguistico

En este punto pasamos de texto plano a texto enriquecido linguisticamente. `spaCy` toma una cadena y devuelve un objeto `Doc` con tokens, lemas, categorias gramaticales y otra informacion util para el analisis.


```python
!pip install spacy -q
!python -m spacy download es_core_news_sm -q
```

```python
import spacy
```

```python
# Cargamos el modelo de espanol
nlp = spacy.load("es_core_news_sm")

# Procesamos el texto
doc = nlp(texto)

print(doc)
```

## 3. Que informacion agrega spaCy al texto

A partir de `doc`, ya no trabajamos solo con caracteres: trabajamos con unidades linguisticas y con atributos asociados a cada una.


**Tokenizacion: la unidad minima de analisis**

Un token es una unidad minima de texto con significado, como una palabra, un numero o un signo de puntuacion.


```python
for token in doc:
    print(token.text)
```

**Lematizacion: la forma base de cada palabra**

El lema es la forma canonica o de diccionario de una palabra. Permite agrupar variantes que comparten significado.


```python
print("PALABRA		LEMA")
print("-------		----")
for token in doc:
    print(f"{token.text:<15}	{token.lemma_}")
```

**Etiquetado gramatical (POS): el rol de cada palabra**

Consiste en asignar a cada palabra su funcion gramatical, por ejemplo sustantivo, verbo o adjetivo.


```python
print("PALABRA		POS		EXPLICACION")
print("-------		---		-----------")
for token in doc:
    print(f"{token.text:<15}	{token.pos_:<10}	{spacy.explain(token.pos_)}")
```
